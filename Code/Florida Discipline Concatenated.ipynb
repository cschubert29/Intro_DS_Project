```python
import pandas as pd
```

```python
#here I'm reading all the discipline csv files we'll be using and concatenating

dis0809df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/0809DisciplineReport.csv",thousands=',')
dis0910df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/0910DisciplineReport.csv",thousands=',')
dis1011df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/1011DisciplineReport.csv",thousands=',',header=2)
dis1112df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/1112DisciplineReport.csv",thousands=',',header=3)
dis1213df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/1213DisciplineReport.csv",thousands=',',header=5)
dis1314df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/1314DisciplineReport.csv",thousands=',',header=4)
dis1415df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/1415DisciplineReport.csv",thousands=',',header=4)
dis1516df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/1516DisciplineReport.csv",thousands=',',header=3)
dis1617df= pd.read_csv("../Data/Florida data/FLDOE/Discipline/1617DisciplineReport.csv",thousands=',',header=3)
```

```python
#Remove columns
dis1314df.drop(['Unnamed: 7','Unnamed: 0'],inplace=True,axis=1)

#Rename columns
dis1314df.rename(columns={"Unnamed: 1": "District"},inplace=True)

#Remove fist row which represents Florida totals
dis1314df = dis1314df.iloc[1:]
dis1314df.reset_index(drop=True,inplace=True)

#Create total column
dis1314df['Total Discipline'] = dis1314df.sum(axis=1)

#Add a year column
dis1314df['Year'] = '2014'

#Remove columns
dis1314df.drop(['Out-of-School','In-School','Unnamed: 4','Without Services','With Services'],inplace=True,axis=1)
```

## Clean up 14-15 file

```python
dis1415df.head()
```

```python
dis1415df.info()
```

```python
#Remove columns
dis1415df.drop(['White','Black or African American','Hispanic/Latino','Asian','Native Hawaiian or Other Pacific Islander'
               ,'American Indian or Alaska Native','Two or More Races','Female','Male','District'],inplace=True,axis=1)

#Rename columns
dis1415df.rename(columns={"District Name": "District", "Total":"Total Discipline"},inplace=True)
```

```python
dis1415df.head(10)
```

```python
#Filter out discipline descriptions: Corporal Punishment, Expelled Without Continuing Educational Services,
#Expelled With Continuing Educational Services, Suspension In-School,Suspension Out-of-School

filter_list = ['Corporal Punishment', 'Expelled Without Continuing Educational Services', 
               'Expelled With Continuing Educational Services','Suspension In-School','Suspension Out-of-School']

dis1415df = dis1415df[dis1415df['Discipline Description'].isin(filter_list)]

#Convert Total column to numeric
dis1415df['Total Discipline'] = pd.to_numeric(dis1415df['Total Discipline'].astype(str).str.replace(',',''), errors='coerce')

#Group by District
dis1415df = dis1415df.groupby(by='District', as_index=False).sum()

#Add a year column
dis1415df['Year'] = '2015'

#Remove Florida row which represents totals
i = dis1415df[dis1415df.District == 'FLORIDA'].index
dis1415df.drop(i,inplace=True)
dis1415df.reset_index(drop=True,inplace=True)
```

```python
dis1415df.head()
```
