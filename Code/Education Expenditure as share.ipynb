# Simple forecasting methods for small dataset
***

* Persistence Forecast
    * Historical mean and median
* Expanding Window Forecast
* Rolling Window Forecast
    * Moving average


```python
#Initialize X, train, test
X = eduexpdf[['Public education expenditure']].values
train, test = X[0:-10], X[-10:]
```

# Persistence Forecast
***

Involves using the previous observation to predict the next time step.

```python
persistence_values = range(1, 10)
scores = list()
for p in persistence_values:
    # walk-forward validation
    history = [x for x in train]
    predictions = list()
    for i in range(len(test)):
        # make prediction
        yhat = history[-p]
        predictions.append(yhat)
        # observation
        history.append(test[i])
    # report performance
    rmse = np.sqrt(mean_squared_error(test, predictions))
    scores.append(rmse)
    print('p=%d RMSE:%.3f' % (p, rmse))
# plot scores over persistence values
plt.plot(persistence_values, scores)
plt.show()
```

```python
#t-1 model

# walk-forward validation
history = [x for x in train]
predictions = list()
for i in range(len(test)):
    # make prediction
    yhat = history[-1]
    predictions.append(yhat)
    # observation
    history.append(test[i])
# plot predictions vs observations
plt.plot(test)
plt.plot(predictions)
plt.show()
```

# Expanding Window Forecast
***
Calculates a statistic on all available historic data and uses that to make a forecast.

```python
#Using the mean
# walk-forward validation
history = [x for x in train]
predictions = list()
for i in range(len(test)):
    # make prediction
    yhat = np.mean(history)
    predictions.append(yhat)
    # observation
    history.append(test[i])
# report performance
rmse = np.sqrt(mean_squared_error(test, predictions))
print('RMSE: %.3f' % rmse)
```

```python
#Using the median
# walk-forward validation
history = [x for x in train]
predictions = list()
for i in range(len(test)):
    # make prediction
    yhat = np.median(history)
    predictions.append(yhat)
    # observation
    history.append(test[i])
# report performance
rmse = np.sqrt(mean_squared_error(test, predictions))
print('RMSE: %.3f' % rmse)
```

```python
#Plot the median expanding window predictions against the test dataset

# walk-forward validation
history = [x for x in train]
predictions = list()
for i in range(len(test)):
    # make prediction
    yhat = np.median(history)
    predictions.append(yhat)
    # observation
    history.append(test[i])
# plot predictions vs observations
plt.plot(test)
plt.plot(predictions)
plt.show()
```

# Rolling Window Forecast
***
Involves calculating a statistic on a rolling block of prior observations and using it as a forecast.


## Moving Average

```python
#Using mean statistic
window_sizes = range(1, 10)
scores = list()
for w in window_sizes:
    # walk-forward validation
    history = [x for x in train]
    predictions = list()
    for i in range(len(test)):
        # make prediction
        yhat = np.mean(history[-w:])
        predictions.append(yhat)
        # observation
        history.append(test[i])
    # report performance
    rmse = np.sqrt(mean_squared_error(test, predictions))
    scores.append(rmse)
    print('w=%d RMSE:%.3f' % (w, rmse))
# plot scores over window sizes values
plt.plot(window_sizes, scores)
plt.show()
```

```python
#Using median statistic
window_sizes = range(1, 10)
scores = list()
for w in window_sizes:
    # walk-forward validation
    history = [x for x in train]
    predictions = list()
    for i in range(len(test)):
        # make prediction
        yhat = np.median(history[-w:])
        predictions.append(yhat)
        # observation
        history.append(test[i])
    # report performance
    rmse = np.sqrt(mean_squared_error(test, predictions))
    scores.append(rmse)
    print('w=%d RMSE:%.3f' % (w, rmse))
# plot scores over window sizes values
plt.plot(window_sizes, scores)
plt.show()
```

### Best window size for either statistic is w=1, which is the same as a t-1 persistence model.

```python

X = eduexpdf[['Public education expenditure']]
train, test = X[0:-10], X[-10:]

# walk-forward validation
history = [x for x in train.values]
predictions = list()
for i in range(len(test.values)):
    # make prediction
    yhat = history[-1]
    predictions.append(yhat)
    # observation
    history.append(test.values[i])
# plot predictions vs observations
plt.figure(figsize=(12,6))
plt.plot(train.index,train.values, label='Train')
plt.plot(test.index,test.values, label='Test')
plt.plot(test.index,predictions, label='Persistence')
plt.xticks(np.arange(min(train.index), max(test.index)+1, 4.0))
plt.legend(loc='best')
plt.title("Persistence t-1 Forecast")
plt.show()
```
