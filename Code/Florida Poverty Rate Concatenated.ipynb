```python
import pandas as pd
import re
import datetime
```

```python
#Reading all the poverty dat/text files we'll be using and concatenating

file_path = "../Data/Florida data/Demographics/Kids in poverty by school district/"
file_list = ['sd99-fl.dat','sd00-fl.dat','sd01-fl.dat','sd02-fl.dat','sd03-fl.txt','sd04-fl.txt','sd05-fl.txt','sd06-fl.txt',
            'sd07-fl.txt','sd08-fl.txt','sd09-fl.txt','sd10-fl.txt','sd11-fl.txt','sd12-fl.txt','sd13-fl.txt',
            'sd14-fl.txt','sd15-fl.txt','sd16-fl.txt']

df_list = []

for f in file_list:
    df_list.append(pd.read_fwf(file_path+f,header=None, 
                    names= ['FIPS State Code','CCD District ID','District Name','Total Population',
                              'Population of Children 5 to 17 years of Age',
                             'Estimated Number of Children in Poverty','File Name','Created Date']))
```

```python
poverty_estimates = pd.concat(df_list, ignore_index=True)
```

```python
poverty_estimates.head()
```

```python
#Function to get the year from the File Name column
def get_year(x):
    y = re.search('\d{2}',x).group(0)
    return datetime.datetime.strptime(y,'%y').year
```

```python
#Add year column
poverty_estimates['Year']=poverty_estimates['File Name'].apply(lambda x: get_year(x))
```

```python
#Add child poverty rate column
#child poverty rate = estimated number of children in poverty/population of children 5-17

poverty_estimates['Child Poverty Rate'] = poverty_estimates['Estimated Number of Children in Poverty']/poverty_estimates['Population of Children 5 to 17 years of Age']
```
