```python
import pandas as pd

#Load data for juvenile populations

juv_pop=pd.read_csv('..\Data\Florida data\Juvenile Crime by County\FL_Juvenile_Pop.csv')
juv_pop.drop([0],axis=0, inplace=True) #Drops all counties
juv_pop.reset_index(drop=True, inplace=True)
juv_pop.head()
```

```python
#Upload excel file with all the juvenile crime data for years 1998 to 2017

xls = pd.ExcelFile(r'..\Data\Florida data\Juvenile Crime by County\UCR_Arrest_Breakdown_County_1998-2017 (2).xlsx')

#create a list of the years to make it easier to load the data into separate data frames

years=[]
for x in range(1998,2018):
    years.append(str(x))

#Read in excel sheet for each year    
FL_juv_crime=pd.read_excel(xls,years)

 


for x in years:
    
    #This section removes unnecessary rows and columns 
    df_dic=FL_juv_crime[str(x)].copy()
    df_dic.drop([0,1,69,70,71],axis=0, inplace=True)
    df_dic = df_dic[['COUNTY NAME','Unnamed: 6', 'Unnamed: 7']].copy()
    
    #Here we calculate total juvenile arrests per county by adding Male and Female arrests
    df_dic['TOTAL ARRESTS']= df_dic['Unnamed: 6'] + df_dic['Unnamed: 7']
    
    #Just keep County Name and the total arrests. 
    df_dic = df_dic[['COUNTY NAME', 'TOTAL ARRESTS']]
    
    #In this section we combine the two data sets for juvenile arrests and juvenile populations and calculate the crime rate
    FL_juv_crime[str(x)] = df_dic
    FL_juv_crime[str(x)].reset_index(drop=True, inplace=True)
    FL_juv_crime[str(x)]['JUVENILE POP'] = juv_pop[str(x)]
    FL_juv_crime[str(x)]['CRIME RATE'] = (FL_juv_crime[str(x)]['TOTAL ARRESTS']/FL_juv_crime[str(x)]['JUVENILE POP'])*100


#This is to verify that the for loop gives us the result we want
#This is the first entry of the dictionary
FL_juv_crime['1998'].head()
```
