```python
#Imports
import pandas as pd
```

```python
#Load data
grades14 = pd.read_csv(r'../Data/Florida data/FLDOE/District Grades/DistrictGrades14.csv',header=2)
grades18 = pd.read_csv(r'../Data/Florida data/FLDOE/District Grades/DistrictGrades18.csv',header=2)
```

```python
grades18.head()
```

```python
#Merge two dataframes
grades_df = pd.merge(grades18,grades14, left_on='District Number', right_on='District Number',how='inner')
```

```python
#list of all columns in merged df
cols = list(grades_df.columns)
```

```python
#Columns to keep
g = ['District Number','District Name_x']
#add columns which contain "Grade" to keep list
g.extend([s for s in cols if 'Grade' in s])
# Select the columns to keep
grades_df1 = grades_df[g]
```

```python
grades_df1.head()
```

```python
#drop columns
grades_df2 = grades_df1.drop(['Did This District Benefit from the One-Letter-Grade-Drop Protection?','Grade 2014_y',
               'Grade 2013_y','Grade 2012_y','Grade 2011_y','Grade 2010_y'],axis=1)

#Rename columns
grades_df2 = grades_df2.rename(columns={'District Name_x':'District Name','Grade 2014_x':'Grade 2014',
                  'Grade 2013_x':'Grade 2013','Grade 2012_x':'Grade 2012',
                  'Grade 2011_x':'Grade 2011','Grade 2010_x':'Grade 2010'})
```

```python
grades_df2.head()
```

```python
#Read grades_df2 to csv

grades_df2.to_csv(r'..\Data\Florida data\FLDOE\District Grades\Grades Concatenated.csv')
```
