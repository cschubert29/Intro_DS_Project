```python
import pandas as pd
```

```python
districts = pd.read_csv(r"..\Data\US-Educational-Finances\districts.csv")
districts.head()
```

```python

districts_FL = districts.loc[districts['STATE']=='Florida']
districts_FL.tail()
```

```python
districts_FL.isna().sum()
```

```python
districts_FL.head()
```

```python
std_districts_FL = districts_FL.iloc[:,4:].div(districts_FL.ENROLL, axis=0)
std_districts_FL = std_districts_FL.join(districts_FL['NAME'])
std_districts_FL = std_districts_FL.join(districts_FL['YRDATA'])
std_districts_FL.head()
```

```python
years=[]
for x in range(1998,2018):
    years.append(str(x))
    
std_districts_FL = std_districts_FL.loc[std_districts_FL['YRDATA'].isin(years)]   

std_districts_FL.tail()
```

```python


std_districts_FL.reset_index(drop=True, inplace=True)
std_districts_FL.head()
```

```python
cols = [ 'NAME',
     'YRDATA',
     'TOTALREV',
     'TFEDREV',
     'TSTREV',
     'TLOCREV',
     'TOTALEXP',
     'TCURINST',
     'TCURSSVC',
     'TCURONON',
     'TCAPOUT']

std_districts_FL = std_districts_FL[cols]
std_districts_FL.head()
```

```python
std_districts_FL.to_csv(r"..\Data\Florida Data\District Finances\FL_District_Finances.csv")
```
